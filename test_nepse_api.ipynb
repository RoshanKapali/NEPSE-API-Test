## 9. Date-Based Data Retrieval

This section demonstrates how to fetch data for specific dates or date ranges.

```python
# Install ipywidgets if not already installed
# !pip install ipywidgets

from datetime import datetime, timedelta, date
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd
```

### 9.1 Interactive Date Picker - Get Floorsheet for a Specific Date

```python
# Create a date picker widget
date_picker = widgets.DatePicker(
    description='Select Date:',
    value=date.today() - timedelta(days=1),  # Default to yesterday
    disabled=False,
    style={'description_width': 'initial'}
)

# Create a symbol input for filtering
symbol_input = widgets.Text(
    description='Stock Symbol (optional):',
    placeholder='e.g., NABIL',
    style={'description_width': 'initial'}
)

# Create a button to fetch data
fetch_button = widgets.Button(
    description='Fetch Data',
    button_style='success',
    tooltip='Click to fetch floorsheet data',
    icon='download'
)

# Output widget to display results
output = widgets.Output()

def on_fetch_clicked(b):
    with output:
        clear_output(wait=True)
        selected_date = date_picker.value
        symbol = symbol_input.value.strip().upper()
        
        print(f"Fetching data for: {selected_date}")
        if symbol:
            print(f"Filtering for symbol: {symbol}")
        
        try:
            if symbol:
                # Fetch floorsheet for specific company
                data = nepse.getFloorSheetOf(symbol, business_date=selected_date)
                print(f"\nFloorsheet records for {symbol} on {selected_date}: {len(data)}")
            else:
                # Fetch all floorsheet data
                print("This may take a moment...")
                # Note: The API might not support historical dates for all floorsheet
                # It primarily works with business_date parameter for company-specific data
                print("\nNote: To fetch all floorsheet for a specific date, use a company symbol.")
                print("The general getFloorSheet() method fetches today's data.")
                data = []
            
            if data:
                # Display first 10 records
                print("\nFirst 10 records:")
                df = pd.DataFrame(data[:10])
                display(df)
                
                # Display summary statistics
                if len(data) > 0:
                    print(f"\nTotal records: {len(data)}")
                    if 'contractAmount' in df.columns:
                        print(f"Total contract amount: {df['contractAmount'].sum():,.2f}")
            else:
                print("\nNo data found for the selected date/symbol.")
                
        except Exception as e:
            print(f"Error: {e}")

fetch_button.on_click(on_fetch_clicked)

# Display widgets
display(widgets.VBox([
    widgets.HBox([date_picker, symbol_input]),
    fetch_button,
    output
]))
```

### 9.2 Date Range Picker - Get Data for Multiple Dates

```python
# Create date range pickers
start_date_picker = widgets.DatePicker(
    description='Start Date:',
    value=date.today() - timedelta(days=7),  # Default to last week
    disabled=False,
    style={'description_width': 'initial'}
)

end_date_picker = widgets.DatePicker(
    description='End Date:',
    value=date.today() - timedelta(days=1),  # Default to yesterday
    disabled=False,
    style={'description_width': 'initial'}
)

# Symbol input for company price/volume history
range_symbol_input = widgets.Text(
    description='Stock Symbol:',
    placeholder='e.g., NABIL',
    value='NABIL',
    style={'description_width': 'initial'}
)

# Data type selector
data_type_selector = widgets.Dropdown(
    options=['Price/Volume History', 'Floorsheet (by date)'],
    value='Price/Volume History',
    description='Data Type:',
    style={'description_width': 'initial'}
)

# Fetch button
fetch_range_button = widgets.Button(
    description='Fetch Range Data',
    button_style='info',
    tooltip='Click to fetch data for date range',
    icon='chart-line'
)

# Output widget
range_output = widgets.Output()

def on_fetch_range_clicked(b):
    with range_output:
        clear_output(wait=True)
        start = start_date_picker.value
        end = end_date_picker.value
        symbol = range_symbol_input.value.strip().upper()
        data_type = data_type_selector.value
        
        if not symbol:
            print("Please enter a stock symbol!")
            return
            
        if start > end:
            print("Start date must be before or equal to end date!")
            return
        
        print(f"Fetching {data_type} for {symbol}")
        print(f"Date Range: {start} to {end}")
        print(f"Total days: {(end - start).days + 1}")
        
        try:
            if data_type == 'Price/Volume History':
                # Use the built-in date range function
                data = nepse.getCompanyPriceVolumeHistory(symbol, start_date=start, end_date=end)
                
                if data and 'content' in data:
                    records = data['content']
                    print(f"\nFound {len(records)} trading days")
                    
                    if records:
                        df = pd.DataFrame(records)
                        
                        # Display summary statistics
                        print("\n=== Summary Statistics ===")
                        if 'closePrice' in df.columns:
                            print(f"Highest Close: {df['closePrice'].max()}")
                            print(f"Lowest Close: {df['closePrice'].min()}")
                            print(f"Average Close: {df['closePrice'].mean():.2f}")
                        
                        if 'totalTradedQuantity' in df.columns:
                            print(f"Total Volume: {df['totalTradedQuantity'].sum():,.0f}")
                            print(f"Average Volume: {df['totalTradedQuantity'].mean():,.0f}")
                        
                        # Display the dataframe
                        print(f"\n=== Price/Volume Data ===")
                        display(df)
                        
                        # Optional: Create a simple plot
                        if 'closePrice' in df.columns and 'businessDate' in df.columns:
                            import matplotlib.pyplot as plt
                            
                            fig, ax1 = plt.subplots(figsize=(12, 6))
                            
                            ax1.set_xlabel('Date')
                            ax1.set_ylabel('Close Price', color='tab:blue')
                            ax1.plot(df['businessDate'], df['closePrice'], color='tab:blue', marker='o')
                            ax1.tick_params(axis='y', labelcolor='tab:blue')
                            ax1.tick_params(axis='x', rotation=45)
                            
                            if 'totalTradedQuantity' in df.columns:
                                ax2 = ax1.twinx()
                                ax2.set_ylabel('Volume', color='tab:orange')
                                ax2.bar(df['businessDate'], df['totalTradedQuantity'], alpha=0.3, color='tab:orange')
                                ax2.tick_params(axis='y', labelcolor='tab:orange')
                            
                            plt.title(f'{symbol} - Price and Volume ({start} to {end})')
                            plt.tight_layout()
                            plt.show()
                else:
                    print("No data available for this date range.")
                    
            else:  # Floorsheet by date
                print("\nFetching floorsheet data for each date...")
                all_data = []
                
                current_date = start
                while current_date <= end:
                    try:
                        daily_data = nepse.getFloorSheetOf(symbol, business_date=current_date)
                        if daily_data:
                            for record in daily_data:
                                record['fetchDate'] = str(current_date)
                            all_data.extend(daily_data)
                            print(f"{current_date}: {len(daily_data)} records")
                        else:
                            print(f"{current_date}: No data (likely non-trading day)")
                    except Exception as e:
                        print(f"{current_date}: Error - {e}")
                    
                    current_date += timedelta(days=1)
                
                if all_data:
                    print(f"\n=== Total Records: {len(all_data)} ===")
                    df = pd.DataFrame(all_data)
                    display(df)
                    
                    # Summary
                    if 'contractAmount' in df.columns:
                        print(f"\nTotal Contract Amount: {df['contractAmount'].sum():,.2f}")
                    if 'contractQuantity' in df.columns:
                        print(f"Total Quantity: {df['contractQuantity'].sum():,.0f}")
                else:
                    print("\nNo floorsheet data found for this date range.")
                    
        except Exception as e:
            print(f"Error: {e}")
            import traceback
            traceback.print_exc()

fetch_range_button.on_click(on_fetch_range_clicked)

# Display widgets
display(widgets.VBox([
    widgets.HBox([start_date_picker, end_date_picker]),
    range_symbol_input,
    data_type_selector,
    fetch_range_button,
    range_output
]))
```

### 9.4 Bulk Data Extraction - Multiple Companies Over Date Range

This example shows how to extract data for multiple companies across a date range:

```python
# Example: Get data for multiple companies
symbols = ["NABIL", "SCB", "NICA"]  # Add your desired symbols
start_date = date(2026, 1, 20)
end_date = date(2026, 1, 30)

all_companies_data = {}

print("Fetching data for multiple companies...")
print(f"Date Range: {start_date} to {end_date}\n")

for symbol in symbols:
    try:
        print(f"Fetching {symbol}...", end=" ")
        data = nepse.getCompanyPriceVolumeHistory(symbol, start_date=start_date, end_date=end_date)
        
        if data and 'content' in data:
            all_companies_data[symbol] = pd.DataFrame(data['content'])
            print(f"✓ ({len(data['content'])} records)")
        else:
            print("✗ No data")
            all_companies_data[symbol] = pd.DataFrame()
    except Exception as e:
        print(f"✗ Error: {e}")
        all_companies_data[symbol] = pd.DataFrame()

# Display summary
print("\n" + "="*80)
print("Summary:")
for symbol, df in all_companies_data.items():
    if not df.empty:
        print(f"\n{symbol}:")
        print(f"  Trading days: {len(df)}")
        if 'closePrice' in df.columns:
            print(f"  Price range: {df['closePrice'].min()} - {df['closePrice'].max()}")
        if 'totalTradedQuantity' in df.columns:
            print(f"  Total volume: {df['totalTradedQuantity'].sum():,.0f}")

# You can also combine all data into a single DataFrame
print("\n" + "="*80)
print("Combined Data:")
combined_df = pd.concat(
    [df.assign(symbol=symbol) for symbol, df in all_companies_data.items() if not df.empty],
    ignore_index=True
)

if not combined_df.empty:
    display(combined_df)
else:
    print("No data to display")
```

### 📝 Key Notes on Date-Based Data Retrieval

**Available Methods:**

1. **`getFloorSheetOf(symbol, business_date)`** - Get floorsheet transactions for a specific company on a specific date
   - Parameters: 
     - `symbol`: Stock symbol (e.g., "NABIL")
     - `business_date`: Date string "YYYY-MM-DD" or date object
   
2. **`getCompanyPriceVolumeHistory(symbol, start_date, end_date)`** - Get price and volume history for a date range
   - Parameters:
     - `symbol`: Stock symbol
     - `start_date`: Start date (date object)
     - `end_date`: End date (date object)
   - Returns: Historical price, volume, and trading statistics

**Important Considerations:**

- ⚠️ Data is only available for **trading days** (market open days)
- 📅 Weekends and holidays will return empty results
- 🔄 The `getFloorSheet()` method (without parameters) fetches **today's data only**
- 💡 For historical data, always use methods with date parameters
- ⏱️ Fetching large date ranges may take time, especially for floorsheet data
- 📊 Use `pandas` DataFrames for easier data analysis and visualization

**Best Practices:**

1. Always handle exceptions when fetching data
2. Check if data is available before processing
3. For large date ranges, consider fetching data in smaller chunks
4. Save fetched data to CSV/Excel to avoid repeated API calls

**What Works:**

1. ✅ **Interactive Date Picker Widget** (Section 9.1)
   - Calendar interface for date selection
   - User-friendly GUI in Jupyter notebooks

2. ✅ **Date Range Picker Widget** (Section 9.2)

**Known Limitations:**

⚠️ **`getFloorSheetOf(symbol, business_date)`** may have API restrictions:
- The endpoint appears to work best with today's/very recent dates
- Historical floorsheet data may not be accessible via this method
- For historical transaction-level data, consider using `getCompanyPriceVolumeHistory()` instead

**Recommendation:**
For historical data analysis, use **`getCompanyPriceVolumeHistory()`** which provides comprehensive daily trading statistics and works reliably across date ranges.

### 🎯 Complete Working Example - Historical Data Analysis

```python
# Complete example: Analyze a stock over a custom date range
print("📊 Historical Data Analysis Tool")
print("=" * 70)

# Configuration - Customize these values
SYMBOL = "NABIL"
START_DATE = date(2026, 1, 15)
END_DATE = date(2026, 1, 30)

print(f"\nAnalyzing: {SYMBOL}")
print(f"Period: {START_DATE} to {END_DATE}")
print(f"Duration: {(END_DATE - START_DATE).days + 1} days\n")

try:
    # Fetch the data
    data = nepse.getCompanyPriceVolumeHistory(SYMBOL, start_date=START_DATE, end_date=END_DATE)
    
    if data and 'content' in data:
        df = pd.DataFrame(data['content'])
        
        if not df.empty:
            # Display key statistics
            print("📈 KEY STATISTICS")
            print("-" * 70)
            print(f"Trading Days: {len(df)}")
            
            if 'closePrice' in df.columns:
                print(f"\nPrice Analysis:")
                first_close = df.iloc[0]['closePrice']
                last_close = df.iloc[-1]['closePrice']
                print(f"  First Closing Price: Rs. {first_close:,.2f}")
                print(f"  Last Closing Price:  Rs. {last_close:,.2f}")
                
                if 'highPrice' in df.columns:
                    print(f"  Highest Price: Rs. {df['highPrice'].max():,.2f}")
                if 'lowPrice' in df.columns:
                    print(f"  Lowest Price:  Rs. {df['lowPrice'].min():,.2f}")
                
                price_change = last_close - first_close
                price_change_pct = (price_change / first_close) * 100
                print(f"  Price Change: Rs. {price_change:,.2f} ({price_change_pct:+.2f}%)")
            
            if 'totalTradedQuantity' in df.columns:
                print(f"\nVolume Analysis:")
                print(f"  Total Volume: {df['totalTradedQuantity'].sum():,.0f} shares")
                print(f"  Average Daily Volume: {df['totalTradedQuantity'].mean():,.0f} shares")
                print(f"  Peak Volume: {df['totalTradedQuantity'].max():,.0f} shares")
            
            if 'totalTrades' in df.columns:
                print(f"\nTransaction Analysis:")
                print(f"  Total Transactions: {df['totalTrades'].sum():,.0f}")
                print(f"  Average Daily Transactions: {df['totalTrades'].mean():,.0f}")
            
            # Display the data table
            print("\n" + "=" * 70)
            print("📋 DETAILED DATA")
            print("-" * 70)
            display(df)
            
            # Optional: Simple visualization
            if 'closePrice' in df.columns and len(df) > 1:
                import matplotlib.pyplot as plt
                
                fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))
                
                # Price chart
                ax1.plot(range(len(df)), df['closePrice'], marker='o', linewidth=2, markersize=6)
                ax1.set_title(f'{SYMBOL} - Closing Price Trend', fontsize=14, fontweight='bold')
                ax1.set_xlabel('Trading Day', fontsize=11)
                ax1.set_ylabel('Price (Rs.)', fontsize=11)
                ax1.grid(True, alpha=0.3)
                ax1.set_xticks(range(len(df)))
                if 'businessDate' in df.columns:
                    ax1.set_xticklabels(df['businessDate'], rotation=45, ha='right')
                
                # Volume chart
                if 'totalTradedQuantity' in df.columns:
                    ax2.bar(range(len(df)), df['totalTradedQuantity'], alpha=0.7, color='orange')
                    ax2.set_title(f'{SYMBOL} - Trading Volume', fontsize=14, fontweight='bold')
                    ax2.set_xlabel('Trading Day', fontsize=11)
                    ax2.set_ylabel('Volume (shares)', fontsize=11)
                    ax2.grid(True, alpha=0.3, axis='y')
                    ax2.set_xticks(range(len(df)))
                    if 'businessDate' in df.columns:
                        ax2.set_xticklabels(df['businessDate'], rotation=45, ha='right')
                
                plt.tight_layout()
                plt.show()
        else:
            print("❌ No data available for this date range.")
    else:
        print("❌ No data returned from API.")
        
except Exception as e:
    print(f"❌ Error: {e}")
    import traceback
    traceback.print_exc()

print("\n" + "=" * 70)
print("✅ Analysis Complete!")
```
